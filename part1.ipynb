```python
import pandas as pd
```

```python
df = pd.read_csv('kc_house_data.csv')
```

```python
df
```

houses built after year 2000

```python
adf = df[df['yr_built']>=2000]
```

```python
adf
```

2. most expensive houses by year

```python
mxHouses = adf.groupby('yr_built')['price'].idxmax()
mxHouses = adf.loc[mxHouses]
mxHouses = mxHouses[['yr_built','price']]
mxHouses
```

3. houses with most bedrooms

```python
mbdr = adf.groupby('yr_built')['bedrooms'].idxmax()
mbdr = adf.loc[mbdr]
mbdr = mbdr[['yr_built','bedrooms','price']]
mbdr
```

4. most bathrooms

```python
mbth = adf.groupby('yr_built')['bathrooms'].idxmax()
mbth = df.loc[mbth]
mbth = mbth[['yr_built','bathrooms','price']]
mbth
```

5. max floors

```python
mflrs = adf.groupby('yr_built')['floors'].idxmax()
mflrs = adf.loc[mflrs]
mflrs = mflrs[['yr_built','price','floors']]
mflrs
```

6 . houses with an area larger than the mean area of all houses

```python
sqftLivingM = adf['sqft_living'].mean()
sqftLivingM = adf[adf['sqft_living']>sqftLivingM]
sqftLivingM = sqftLivingM[['sqft_living','price']]
sqftLivingM.head(10)
```

houses with area larger than the mean of the sqft_lot

```python
sqftLotM = adf['sqft_lot'].mean()
sqftLotM = adf[adf['sqft_lot']>sqftLotM]
sqftLotM = sqftLotM[['sqft_lot','price']]
sqftLotM.head(10)
```
